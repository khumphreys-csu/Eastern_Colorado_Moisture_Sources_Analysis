# Script to create a CSV contain information about all source regions

Rows: One region

Columns: Characteristics attributing to each source region.
Including:
- 'region_name' : Full region name for titles and maps
- 'file_name' : Full region name with underscores intead of spaces for file naming 
- 'short_name' : Abreviated names used for the legends/keys
- 'other_acronym': CO alt climate divisions have two names, this is their other acronym
- 'surface_type': Whether the region is considered land or water
- 'clim_division': Boolean of whether a region is a colorado climate division or not
- 'mask_filepath': path to specific mask for the region
